```python
Total_run=sum(data['runs'])
print('Total Run = ',Total_run)
```

```python
Total_odi_match_innings=0
Total_T20_match_innings=0
Total_Test_match_innings=0
for i in data['match']:
  if i=='ODI':
    Total_odi_match_innings+=1
  if i=='T20':
    Total_T20_match_innings+=1
  if i=='Test':
    Total_Test_match_innings+=1
print('Total_odi_match_innings ',Total_odi_match_innings)
print('Total_T20_match_innings ',Total_T20_match_innings)
print('Total_Test_match_innings ',Total_Test_match_innings)
```

```python
#Run Against Pak
run=0
match=0
for i in range(data.shape[0]):
  if data['opponent'][i]=='Pakistan':
    run+=data['runs'][i]
    match+=1
print('Run against Pakistan',run,'in',match,'matches')
```

```python
l=list(set(data['opponent']))
print('Opponents = \n')
l
```

```python
Runs=[]
```

```python
def Run_against_particular_team(teams):
  run=0
  match=0
  for i in range(data.shape[0]):
    if data['opponent'][i]==teams:
      run+=data['runs'][i]
      match+=1
  Runs.append(run)
  print('Total Run againsts',teams,'teams',run,'in',match,'matches')
```

```python
for country in l:
  Run_against_particular_team(country)
```

```python
p=list(set(data['ground']))
print('Ground = \n')
p
```

```python
def Runs_on_particular_Ground(Ground):
  run=0
  match=0
  for i in range(data.shape[0]):
    if data['ground'][i]==Ground:
      run+=data['runs'][i]
      match+=1
  print('Total Runs on',Ground,'ground',run,'in',match,'matches')
```

```python
for ground in p:
  Runs_on_particular_Ground(ground)
```

```python
def Runs_in_particular_year(year):
  run=0
  match=0
  for i in range(data.shape[0]):
    if data['date'][i][-4:]==year:
      run+=data['runs'][i]
      match+=1
  print('Total Runs in',year,'year',run,'in',match,'matches')
```

```python
for year in range(2008,2023):
  Runs_in_particular_year(str(year))
```

```python
city_counts = data['opponent'].value_counts().to_frame().sort_values('opponent', axis=0, ascending=False).head(10)

fig, ax = plt.subplots(figsize=(12, 6))

ax.bar(city_counts.index, city_counts['opponent'], color=['cyan'], width=0.5)

ax.set_title('Number of matches', fontsize=16)
ax.set_xlabel('Opponents', fontsize=14)
ax.set_ylabel('Number of matches', fontsize=14)

ax.tick_params(axis='both', labelsize=12)
ax.set_xticklabels(city_counts.index, rotation=45, ha='right')

for i, v in enumerate(city_counts['opponent']):
    ax.text(i - 0.1, v + 50, str(v), fontsize=12)

ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.6)

plt.show()
```
